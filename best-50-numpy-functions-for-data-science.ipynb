# Array Creation Methods

## 1. Array
It is used to create one-dimensional or multidimensional arrays from scratch.<br>
`array(Object, dtype=None, *, copy=True, order='K', subok=False, ndmin=0)`


**Important params**<br>
`dtype`: desired data type for the resultant array.<br>
`ndim`: Specify the minimum number of dimensions for the resultant array.

````python
import numpy as np
np.array([1,2,3,4,5])
````

You can also use this function to convert series or dataframes into a NumPy array.

````python
import pandas as pd
sex = pd.Series(['Male','Female','Female'])
np.array(sex)
````

### 2. Linspace
Creates an array with evenly spaced float numbers over a specified interval.<br>
`np.linspace(start, stop, num = 50, endpoint=True, retstep=False, dtype=None,axis=0)`

**Important params**<br>
`start`: starting index<br>
`end`: last index<br>
`num`: number of samples to generate, default = 50.

````python
np.linspace(10,100,11)
````

### 3. Arange
Return evenly spaced integer values over a given interval with some step size.<br>
`np.arrange(star, stop, step, dtype=None)`

**Important params**<br>
`step`: Space between values.

````python
np.arange(1,10,3)
````

### 4. Uniform Samples
Generate a random sample from a uniform distribution between lower and higher limit values.<br>
`np.uniform(low=0.0, high=1.0, size=None)`

````python
np.random.uniform(5,10,size=4)
````

````python
np.random.uniform(size=5)
````

````python
np.random.uniform(size = (2,3))
````

### 5. Random.randint
Generate n random integer samples within a range.<br>
`np.random.randiant(low, high=None, size=None, dtype=int)`

````python
np.random.randint(5,15,10)
````

````python
np.random.random(5)
````

### 7. Logspace
Generate evenly spaced numbers on a log scale.<br>
`np.logspace(start, stop, num=50, endpoint=True, base=10.0, dtype=None, axis=0)`

**Important params**<br>
`start`: starting value of the sequence.<br>
`end`: last value of the sequence.<br>
`endpoint`: if True, the last sample will be included in the sequence.<br>
`base`: The base of the log space. default is 10.

````python
np.logspace(0,10,5,base=2)
````

### 8. Array of zero
np.zeroes function helps to create an array of 0<br>
`np.zero(shape, dtype=float, order='C')`

**Important params**<br>
`shape`: shape of the resultant array.<br>
`dtype`: desired data type of resultant array. ‘int’ or default ‘float’

````python
np.zeros((2,3), dtype=int)
````

````python
np.zeros(5)
````

### 9. Array of Ones
np.ones function helps to create an array of 1.<br>
`np.ones(shape, dtype=None, order='C')`

````python
np.ones((3,5))
````

### 10. Array of k Random Value
Creates an n-dimensional array of a random value.<br>
`np.full(shape, fill_value, dtype=None)`<br>

**Important params**<br>
`fill_value`: Random value to fill inside the array.

````python
np.full((2,4), fill_value=2)
````

### 11. Identity
Create an identity matrix with specified dimensions.<br>
`np.identity(n, dtype=None)`

````python
np.identity(4)
````

# Array Operations

### 12. Min
Return the minimum value from the array.<br>
`np.min(a, axis=None, out=None, keepdims=<no value>, initial=<no value>, where=<no value>)`

**Important params**<br>
`axis`: axis along which to operate.<br>
`out`: alternative array to store the output.

````python
arr = np.array([1,1,2,3,4,4,5,5,6,6,2,2])
print(np.min(arr))
````

### 13. Max
Return the maximum value from the array.<br>
`np.max(a, axis=None, out=None)`

````python
np.max(arr)
````

### 14. Unique
Return an array with all the unique elements sorted.<br>
`np.unique(arr, return_index=False, return_inverse=False, return_counts=False, axis=None)`

**Important params**<br>
`return_index`: if True, return the indices of the array.<br>
`return_inverse`: if True, return the indices of the unique array.<br>
`return_counts`: if True, return the number of times each unique element appears inside the array.<br>
`axis`: The axis to operate on. By default, the array is considered flattened.

````python
np.unique(arr, return_counts=True)
````

### 15. Mean
It is used to get the mean of the array.<br>
`np.mean(a, axis=None, dtype=None, out=None)`

````python
np.mean(arr, dtype=int)
````

### 16. Median
Return the median value of the array.<br>
`np.median(a, axis=None, out=None)`

````python
arr = np.array([[1,2,3],[5,9,5]])
np.median(arr)
````

### 17. Digitize
Return the indices of the bins to which each value in the input array belongs.<br>
`np.digitize(x, bins, right=False)`

**Important params**<br>
`bins`: Array of bins.<br>
`right`: Indicates whether the interval includes the right or left bin edge.

````python
a = np.array([-0.9, 0.5, 0.9, 1, 1.2, 1.4, 3.6, 4.7, 5.7])
bins = np.array([0,1,2,3])
np.digitize(a,bins)
````

```
Exp        Value
x < 0     :   0
0 <= x <1 :   1
1 <= x <2 :   2
2 <= x <3 :   3
3 <=x     :   4
Compares -0.9 to 0, here x < 0 so Put 0 in resulting array.
Compares  0.5 to 0, here 0 <= x <1 so Put 1.
.
Compares 5.4 to 4, here 3<=x so Put 4
```

### 18. Reshape
It is one of the most used functions of NumPy. It Returns an array containing the same data with a new shape.<br>
`np.reshape(shape)`

````python
a = np.random.randint(12, size=(4,3))
a
````

````python
a.reshape(3,4)
````

````python
a.reshape(-1)
````

### 19. Expand Dimensions
It is used to expand the dimensions of an array. This method is useful for creating sample test data for testing a machine learning model.<br>
`np.expand_dims(a, axis)`

````python
arr = np.array([8, 14, 15, 1, 6, 11, 4, 9, 4, 12, 13, 13, 11])
np.expand_dims(arr, axis=0)
````

````python
np.expand_dims(arr, axis=1)
````

## 20. Squeeze
Reduce the dimension of an array by removing single-dimensional entrie.<br>
`np.squeeze(a, axis=None)`

````python
arr = np.array([[8],[14],[1],[8],[11],[4],[9],[4],[1],[13],[13],[11]])
np.squeeze(arr)
````
